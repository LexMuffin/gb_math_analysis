1. Исследовать функции на условный экстремум: \
    a) $ U = 3 - 8x + 6y $, если $ x^2 + y^2 = 36 $ \

a) $ U = 3 - 8x + 6y $, если $ x^2 + y^2 = 36 $

$ L(\lambda_1, x, y) = 3 - 8x + 6y + \lambda_1 * (x^2 + y^2 - 36) $

$
\left\{
    \begin{array} \\
        L'_x = -8 + 2x * \lambda_1 = 0 \\
        L'_y = 6 + 2y * \lambda_1 = 0 \\
        L'_{\lambda_1} = x^2 + y^2 - 36 = 0
    \end{array}
\right.
\iff
\left\{
    \begin{array} \\
        x = \frac{4}{\lambda_1} \\
        y = \frac{-3}{\lambda_1} \\
        \frac{16}{\lambda_1^2} + \frac{9}{\lambda_1^2} = 36
    \end{array}
\right.
\iff
\left\{
    \begin{array} \\
        x = \frac{4}{\lambda_1} \\
        y = \frac{-3}{\lambda_1} \\
        \lambda_1^2 = \frac{25}{36}
    \end{array}
\right.
\iff
\left\{
    \begin{array} \\
        x = \pm \frac{24}{5} \\
        y = \mp \frac{18}{5} \\
        \lambda_1 = \pm \frac{5}{6}
    \end{array}
\right.
$

Стационарные точки: $ (\frac{5}{6}, \frac{24}{5}, -\frac{18}{5}), (-\frac{5}{6}, -\frac{24}{5}, \frac{18}{5}) $.

$
L''_{xx} = 2\lambda_1, \quad L''_{yy} = 2\lambda_1, \quad L''_{\lambda_1 \lambda_1} = 0 \\
L''_{xy} = 0, \quad L''_{x \lambda_1} = 2x, \quad L''_{y \lambda_1} = 2y
$

$
A
=
\begin{pmatrix}
    L''_{\lambda_1 \lambda_1} & L''_{\lambda_1 x} & L''_{\lambda_1 y} \\
    L''_{x \lambda_1} & L''_{xx} & L''_{xy} \\
    L''_{y \lambda_1} & L''_{yx} & L''_{yy}
\end{pmatrix}
=
\begin{pmatrix}
    0  & 2x         & 2y \\
    2x & 2\lambda_1 & 0 \\
    2y & 0          & 2\lambda_1
\end{pmatrix} \\
$

$
|A|
=
\begin{vmatrix}
    0  & 2x         & 2y \\
    2x & 2\lambda_1 & 0 \\
    2y & 0          & 2\lambda_1
\end{vmatrix}
=
0 * \begin{vmatrix} 2\lambda_1 & 0 \\ 0 & 2\lambda_1 \end{vmatrix}
- 2x * \begin{vmatrix} 2x & 0 \\ 2y & 2\lambda_1 \end{vmatrix}
+ 2y * \begin{vmatrix} 2x & 2\lambda_1 \\ 2y & 0 \end{vmatrix}
= 0 - 2x * (2x * 2\lambda_1 - 0) + 2y * (0 - 2y * 2\lambda_1)
= -8\lambda_1(x^2 + y^2)
$

$
|A|(\frac{5}{6}, \frac{24}{5}, -\frac{18}{5}) = -8 * \frac{5}{6} * (x^2 + y^2) < 0 \Rightarrow точка \ минимума.\\
|A|(-\frac{5}{6}, -\frac{24}{5}, \frac{18}{5}) = -8 * -\frac{5}{6} * (x^2 + y^2) > 0 \Rightarrow точка \ максимума.
$


b) $ U = 2x^2 + 12xy + 32y^2 + 15 $, если $ x^2 + 16y^2 = 64 $

$ L(\lambda_1, x, y) = 2x^2 + 12xy + 32y^2 + 15 + \lambda_1 * (x^2 + 16y^2 - 64) $
$
\left\{
    \begin{array} \\
        L'_x = 4x + 12y + 2x * \lambda_1 = 0 \\
        L'_y = 64y + 12x + 32y * \lambda_1 = 0 \\
        L'_{\lambda_1} = x^2 + 16y^2 - 64 = 0
    \end{array}
\right.
\iff
\left\{
    \begin{array} \\
        x =  \frac{-6y}{\lambda_1 + 2} \\
        y =  \frac{-3x}{8(\lambda_1 + 2)} \\
        x^2 + 16y^2 - 64 = 0
    \end{array}
\right.
\iff
\left\{
    \begin{array} \\
        \lambda_1 + 2 = \frac{-6y}{x} = \frac{-3x}{8y} \\
        y = \frac{-3x}{8(\lambda_1 + 2)} \\
        x^2 + 16y^2 - 64 = 0
    \end{array}
\right.
\iff
\left\{
    \begin{array} \\
        \lambda_1 + 2 = \frac{-6y}{x} \\
        y = \frac{-3x}{8(\lambda_1 + 2)} \\
        2x^2 = 64
    \end{array}
\right.
\iff
\left\{
    \begin{array} \\
        \lambda_1 = -\frac{6 * \pm \sqrt{2})}{\pm 4\sqrt{2}} - 2 = -1.5 - 2 = -3.5\\
        x = \pm 4\sqrt{2} \\
        y = \pm \sqrt{2}
    \end{array}
\right.
$

Стационарные точки: $ (-3.5, 4\sqrt{2}, \sqrt{2}), (-3.5, -4\sqrt{2}, -\sqrt{2}) $.

$
L''_{xx} = 4 + 2\lambda_1, \quad L''_{yy} = 64 + 32\lambda_1, \quad L''_{\lambda_1 \lambda_1} = 0 \\
L''_{xy} = 12, \quad L''_{x \lambda_1} = 2x, \quad L''_{y \lambda_1} = 32y
$

$
A
=
\begin{pmatrix}
    L''_{\lambda_1 \lambda_1} & L''_{\lambda_1 x} & L''_{\lambda_1 y} \\
    L''_{x \lambda_1} & L''_{xx} & L''_{xy} \\
    L''_{y \lambda_1} & L''_{yx} & L''_{yy}
\end{pmatrix}
=
\begin{pmatrix}
    0   & 2x             & 32y \\
    2x  & 4 + 2\lambda_1 & 12 \\
    32y & 12             & 64 + 32\lambda_1
\end{pmatrix} \\
$

$
|A|
=
\begin{vmatrix}
    0   & 2x             & 32y \\
    2x  & 4 + 2\lambda_1 & 12 \\
    32y & 12             & 64 + 32\lambda_1
\end{vmatrix}
=
0 * \begin{vmatrix} 4 + 2\lambda_1 & 12 \\ 12 & 64 + 32\lambda_1 \end{vmatrix}
- 2x * \begin{vmatrix} 2x & 12 \\ 32y & 64 + 32\lambda_1 \end{vmatrix}
+ 32y * \begin{vmatrix} 2x & 4 + 2\lambda_1 \\ 32y & 12 \end{vmatrix} = \\
= 0 - 2x * (2x * (64 + 32\lambda_1) - 12 * 32y) + 32y * (2x * 12 - 32y * (4 + 2\lambda_1))
= -2x(128x + 64x \lambda_1 - 384y) + 32y(24x - 128y - 64y \lambda_1)
= -256x^2 - 128x^2 \lambda_1 + 768xy + 768xy - 4096y - 2048y^2 \lambda_1
= -128(2x^2 + 32y^2 + 12xy - x^2 \lambda_1 - 16y^2 \lambda_1)
= -128(x^2(2 - \lambda_1) + 16y^2(2 - \lambda_1) + 12xy)
= -128((2 - \lambda_1)(x^2 + 16y^2) + 12xy)
= -128(2 - \lambda_1)(x^2 + 16y^2) - 1536xy
$

$
|A|(-3.5, 4\sqrt{2}, \sqrt{2}) = -128(2 + 3.5)(32 + 32) - 1536*8 = -7 * 2^7 * 2^6 - 3 * 2^9 * 2^3 < 0 \Rightarrow точка \ минимума.\\
|A|(-3.5, -4\sqrt{2}, -\sqrt{2}) = -128(2 + 3.5)(32 + 32) - 1536*8 = -7 * 2^7 * 2^6 - 3 * 2^9 * 2^3 < 0 \Rightarrow точка \ минимума.
$


2. Найти производную функции $ U = x^2 + y^2 + z^2 $ по направлению вектора $ \vec{c}(-9, 8, -12) $ в точку $ M(8, -12, 9) $.

$
U'_x = 2x, \ U'_x(8, -12, 9) = 16 \\
U'_y = 2y, \ U'_y(8, -12, 9) = -24 \\
U'_z = 2z, \ U'_z(8, -12, 9) = 18 \\
|\vec{c}| = \sqrt{9^2 + 8^2 + 12^2} = \sqrt{81 + 64 + 144} = \sqrt{289} = 17 \\
\vec{c_0} = (\frac{-9}{17}, \frac{8}{17}, \frac{-12}{17}) \\
 \frac{\partial U}{\partial c} \Big |_M
= -16 * \frac{9}{17} - 24 * \frac{8}{17} - 18 * \frac{12}{17}
= \frac{-144 - 192 - 216}{17}
= -\frac{552}{17}
= -32\frac{8}{17}
$

3. Найти производную функции $ U = \large e^{x^2 + y^2 + z^2} $ по направлению вектора $ \vec{d}(4, -13, -16) $ в точку $ M(-16, 4, -13) $.

$
U'_x = 2x*e^{x^2 + y^2 + z^2}, \ U'_x(-16, 4, -13) = -32e^441 \\
U'_y = 2y*e^{x^2 + y^2 + z^2}, \ U'_y(-16, 4, -13) = 8e^441 \\
U'_z = 2z*e^{x^2 + y^2 + z^2}, \ U'_z(-16, 4, -13) = -26e^441 \\
|\vec{d}| = \sqrt{4^2 + 13^2 + 16^2} = \sqrt{16 + 169 + 256} = \sqrt{441} = 21 \\
\vec{d_0} =  (\frac{4}{21}, \frac{-13}{21}, \frac{-16}{21}) \\
 \frac{\partial U}{\partial d} \Big |_M
= -32e^441 * \frac{4}{21} - 8e^441 * \frac{13}{21} + 26e^441 * \frac{16}{21}
= \frac{-128e^441 - 104e^441 - 416e^441}{21}
= -\frac{648}{21}
= -30\frac{6}{7}
$
